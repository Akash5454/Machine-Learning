# Kaggle Facebook Dataset Analysis
This exploratory data analysis gives insights from Facebook dataset which consists of identifying users that can be focused more to increase the business.
These valuable insights should help Facebook to take intelligent decision to identify its useful users and provide correct recommendations to them.

This dataset contains 99903 entries with 15 columns.

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading facebook dataset
dataset = pd.read_csv(r"C:\Users\ASUS\Downloads\pseudo_facebook.csv")
dataset.head()
```

```python
# for concise summary of a DataFrame
dataset.info()
# we can see that gender and tenure has missing values 
```

```python
# for descriptive statistics
dataset.describe()
# we observe that 
    # average age of user is 37
    # youngest user is 13 years old whereas oldest user age is 113  
```

```python
# since there are many user who didn't disclosed there gender we have to replace the missing values to unknown for further analysis 
dataset['gender'].fillna("unknown", inplace=True)
```

```python
# dropping other missing values
dataset.dropna(inplace=True)
```

```python
dataset.info()
```

```python
# gender distrinbution of users
dataset['gender'].value_counts()
```

```python
# age-wise distribution of users
dataset['age'].value_counts()
```

```python
# for further analysis we are going to add 'age band' col which puts every user into an age age interval 
dataset['age band'] = pd.cut(dataset['age'], bins=np.arange(10,111,10), labels=['10-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100', 'Above 100'], include_lowest=True)
```

```python
# we can observe that more than half of the users belong to the age group of 10 to 30 years  
dataset['age band'].value_counts()
```

```python
# above analysis can be clearly observed through pie chart which tells 54.1% of users are of 10 to 30 years old
plt.figure(figsize=(10,10)) 
dataset['age band'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
# Gender distribution suggest that most users are male
plt.figure(figsize=(9,9))
dataset['gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
test=dataset.groupby("age band")['tenure'].count()
test=test/sum(test)*100
```

```python
test.index
```

```python
plt.figure(figsize=(10,10))
color=["royalblue","cornflowerblue","steelblue","lightblue","teal"]
plt.bar(test.index,test.values, color=color)
plt.xlabel("Age Band")
plt.ylabel("Percentage of Users")
plt.title("Age Distribution of Users")
```

```python
# We can observe that female likes more content of facebook 
sns.barplot(x = dataset['gender'], y = dataset['likes'])
```

```python
# We can observe that female have more friends
sns.barplot(x = dataset['gender'], y = dataset['friend_count'])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = dataset['age band'], y = dataset['friend_count'], hue=dataset['gender'])
plt.title("No of friends per age group")
```

```python
# for extracting no. of users who have zero friends
zerofriends = dataset[dataset['friend_count']==0]
```

```python
# We can observe that there is more no. of males that have zero friends
zerofriends['gender'].value_counts().plot(kind='bar', color=color[:2])
```

```python
plt.figure(figsize=[7, 7])
zerofriends['age band'].value_counts().plot(kind='bar')
plt.ylabel("No. of Users")
plt.xlabel("Age Group")
plt.title("No. of Users with no friends per age group")
```

```python
# adiing 'tenure_yearly' col to know how much time has been before they get registered on facebook
dataset["tenure_yearly"]=dataset["tenure"].apply(lambda x:x/365)
```

```python
# for further analysis adding a grouping every user into tenure group
labels=['0-1 year','1-2 years','2-3 years','3-4 years','4-5 years','5-6 years','over 6 years']

dataset['tenure_group']=pd.cut(dataset["tenure_yearly"],bins=[0,1,2,3,4,5,6,9],labels=labels,include_lowest=True)
```

```python
dataset.head()
```

```python
# we can observe that most users are new to facebook
dataset.groupby("tenure_group")["tenure_yearly"].count()
```

```python
#  retainsion rate decreases steadly. Almost all users do not use facebook after 5 years
dataset.groupby("tenure_group")["tenure_yearly"].count().plot()
plt.ylabel("No. of Users")
```

```python
dataset.info()
```

```python
# extracting no. of likes per day by every user whose tenure is above 0
dataset['likes_per_day'] = dataset['likes_received']/dataset['tenure'].where(dataset['tenure']>0)
```

```python
dataset.sort_values('likes_received', ascending=False)[:10]
```

```python
# top ten users who recieved maximum likes per day 
dataset.sort_values('likes_per_day', ascending=False)[:10]
```

```python
famousPeople = dataset.sort_values('likes_per_day', ascending=False)[:10]
famousPeople.plot(x='userid',y='likes_per_day',kind='bar')
plt.ylabel("Likes per day")
plt.xlabel("User ID")
plt.title("Top 10 Famous Users")
```

```python
# top 10 users who sends friend requests
MostfriendsInitiated = dataset.sort_values('friendships_initiated', ascending=False)[:10]
```

```python
MostfriendsInitiated['freiendsInitiated_per_day'] =  MostfriendsInitiated['likes_received']/MostfriendsInitiated['tenure'].where(MostfriendsInitiated['tenure']>0)
```

```python
MostfriendsInitiated
```

```python
MostfriendsInitiated.sort_values('freiendsInitiated_per_day', ascending=False).plot(x='userid',y='freiendsInitiated_per_day',kind='bar')
plt.ylabel("Friends Initiated Per Day")
plt.xlabel("User ID")
plt.title("Top 10 Users who sends maximum friend requests")
```
